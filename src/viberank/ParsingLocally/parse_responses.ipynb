```python
import json
from pathlib import Path
import pandas as pd

jsonl_path = Path(r"D:\Vtech\vibeRank\responses\Vispdat_qwen_run.jsonl")

records = []
with open(jsonl_path, "r", encoding="utf-8") as f:
    for line_num, line in enumerate(f, start=1):
        line = line.strip()
        if not line:
            continue
        try:
            records.append(json.loads(line))
        except json.JSONDecodeError as e:
            print(f"Skipping bad JSON on line {line_num}: {e}")

print(f"Loaded {len(records)} records")

# Optional: inspect event types
event_counts = {}
for r in records:
    event = r.get("event", "UNKNOWN")
    event_counts[event] = event_counts.get(event, 0) + 1

print("Event counts:", event_counts)

# Convert to DataFrame for easier viewing
df = pd.DataFrame(records)

# Show the first few rows
display(df.head())

# Show only response records
responses_df = df[df["event"] == "response"].copy()
print(f"Number of response records: {len(responses_df)}")

display(
    responses_df[
        [
            "tie_index",
            "repeat_index",
            "pair_key",
            "raw_response",
            "error",
            "latency_ms",
        ]
    ].head(10)
)
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

MODEL_NAME = "Qwen/Qwen2.5-0.5B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype="auto",
    device_map="auto",
)

if tokenizer.pad_token_id is None and tokenizer.eos_token_id is not None:
    tokenizer.pad_token = tokenizer.eos_token


def build_messages(raw_response: str):
    return [
        {
            "role": "system",
            "content": "Return exactly one label: 1, 2, or U.",
        },
        {
            "role": "user",
            "content": (
                "Classify the text.\n\n"
                "Labels:\n"
                "1 = Household 1 got Emergency Shelter and Household 2 got Transitional Housing.\n"
                "2 = Household 2 got Emergency Shelter and Household 1 got Transitional Housing.\n"
                "U = unclear.\n\n"
                f"Text:\n{raw_response}\n\n"
                "Label:"
            ),
        },
    ]


def parse_one_by_scoring(raw_response: str):
    messages = build_messages(raw_response)

    prompt = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True,
    )

    inputs = tokenizer(prompt, return_tensors="pt", truncation=True).to(model.device)

    with torch.no_grad():
        outputs = model(**inputs)

    next_token_logits = outputs.logits[0, -1, :]

    # Candidate labels as token IDs
    # Use versions with a leading space too, since tokenization can differ
    candidates = {
        "1": ["1", " 1"],
        "2": ["2", " 2"],
        "U": ["U", " U"],
    }

    label_scores = {}

    for label, variants in candidates.items():
        scores = []
        for variant in variants:
            token_ids = tokenizer.encode(variant, add_special_tokens=False)
            if len(token_ids) == 1:
                scores.append(next_token_logits[token_ids[0]].item())
        label_scores[label] = max(scores) if scores else float("-inf")

    best_label = max(label_scores, key=label_scores.get)

    return {
        "original_text": raw_response,
        "label": best_label,
        "scores": label_scores,
        "prompt": prompt,
    }


# Try a few
for x in responses_df["raw_response"].dropna().head(5):
    result = parse_one_by_scoring(x)
    print("=" * 100)
    print("ORIGINAL TEXT:")
    print(result["original_text"])
    print()
    print("LABEL:", result["label"])
    print("SCORES:", result["scores"])
    print()
```

```python
def debug_one(raw_response: str):
    messages = build_messages(raw_response)

    prompt = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True,
    )

    inputs = tokenizer(prompt, return_tensors="pt", truncation=True).to(model.device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=10,
        do_sample=False,
        pad_token_id=tokenizer.pad_token_id,
        eos_token_id=tokenizer.eos_token_id,
        return_dict_in_generate=True,
    )

    decoded_full = tokenizer.decode(outputs.sequences[0], skip_special_tokens=False)
    decoded_clean = tokenizer.decode(outputs.sequences[0], skip_special_tokens=True)

    print("PROMPT LEN:", len(prompt))
    print("PROMPT TAIL:", repr(prompt[-300:]))
    print("FULL DECODE:", repr(decoded_full))
    print("CLEAN DECODE:", repr(decoded_clean))
    print("NEW TOKEN IDS:", outputs.sequences[0][inputs["input_ids"].shape[1]:].tolist())
```
